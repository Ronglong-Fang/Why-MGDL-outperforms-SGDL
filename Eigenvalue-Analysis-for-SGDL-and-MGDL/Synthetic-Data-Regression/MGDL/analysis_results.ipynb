```python
import pickle

path = 'results/*.pickle'
# path = 'results/BEST/*.pickle'

picklefile = []
for file in glob.glob(path):  
    picklefile.append(file)
    # print(file)
print(len(picklefile))
k=4
picklefile =sorted(picklefile, key=lambda x: float(x.split('_')[k][len("trainloss"):]), reverse=False)


learning_rate = []
train_loss = []
val_loss = []
test_loss = []
for file in picklefile:
    print(file)
    with open(file, 'rb') as f:
        [SaveHistory, opt] = pickle.load(f)
    learning_rate.append(opt.learning_rate)
    train_loss.append(SaveHistory['grade'+str(opt.grade)]['train_loss'][-1])
    val_loss.append(SaveHistory['grade'+str(opt.grade)]['val_loss'][-1])
    test_loss.append(SaveHistory['grade'+str(opt.grade)]['test_loss'])

print('\n\n')
```
